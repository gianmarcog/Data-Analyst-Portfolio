```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.distplot(tips["total_bill"], kde=False, bins=10)
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="kde")
```

```python
sns.pairplot(tips)
```

#### Barplot

```python
sns.barplot(x="sex",y="total_bill",data=tips, estimator=np.std)
```

#### Countplot

```python
sns.countplot(x="sex", data=tips)
```

#### Boxplot

```python
sns.boxplot(data=tips,palette="rainbow", orient="h")
```

```python
sns.boxplot(x="day",y="total_bill",hue="smoker",data=tips,palette="coolwarm")
```

#### Violinplot

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", split=True, palette="Set1")
```

#### Stripplot

```python
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True, hue='sex', palette="Set1", dodge=True)
```

#### Swarmplot

```python
sns.swarmplot(x="day", y="total_bill", hue="sex", palette="Set1", dodge=True, s=4, data=tips)
```

```python
sns.violinplot(x='tip', y='day', data=tips, palette='rainbow')
sns.swarmplot(x='tip', y='day', data=tips, color='black', size = 2)
```

#### Catplot ( In the past = Factorplot)

```python
sns.catplot(x='sex', y='total_bill', data=tips, kind='bar')
```
